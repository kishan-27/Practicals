## Practical No.9: Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of given company between two specific dates.

```python
# Name : Kishan B. Lashkari
# Enrollment : 21627307002

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/RELIANCE.csv')

stk_pr = df[0:15]

start_date = '2000-01-03'
end_date = '2000-03-13'

plt.figure() 

up = stk_pr[stk_pr.Close >= stk_pr.Open] 
down = stk_pr[stk_pr.Close < stk_pr.Open] 
col1 = 'blue'
col2 = 'darkgreen'
width = .3
width2 = .03
plt.bar(up.index, up.Close-up.Open, width, bottom=up.Open, color=col1) 
plt.bar(up.index, up.High-up.Close, width2, bottom=up.Close, color=col1) 
plt.bar(up.index, up.Low-up.Open, width2, bottom=up.Open, color=col1) 
plt.bar(down.index, down.Close-down.Open, width, bottom=down.Open, color=col2) 
plt.bar(down.index, down.High-down.Open, width2, bottom=down.Open, color=col2) 
plt.bar(down.index, down.Low-down.Close, width2, bottom=down.Close, color=col2)

plt.title('Stock Prices between {} and {}'.format(start_date, end_date))
plt.xlabel('Stock')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

plt.xticks(rotation=30, ha='right') 

plt.show()
```
