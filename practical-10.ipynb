## Practical No.10: Write a Pandas program to implement following operation
### ● to find and drop the missing values from the given dataset

 

```python
import pandas as pd 

df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/RELIANCE.csv')

missing = df.isna()

df_clean = df.dropna(axis=0)

print(df)
print("\n\n\n\n\n")
print(missing)

print("\n\n\n\n\n\n")
print(df_clean)
```

## ● to remove the duplicates from the given dataset.

```python
# Name : Kishan B. Lashkari
# Enrollment : 216270307002

import pandas as pd

df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/RELIANCE.csv')

df_cleaned = df.drop_duplicates()

print(df)
print("\n\n\n\n\n")
print(df_cleaned)
```
