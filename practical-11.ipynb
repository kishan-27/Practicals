## Practical No.11: Write a Pandas program to filter all columns where all entries present, check which rows and columns has a NaN and finally drop rows with any NaNs from the given dataset.


```python
# Name : Kishan B. Lashkari
# Enrollment : 216270307002

import pandas as pd

df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/RELIANCE.csv') 
print("Reliance Sample Data :")

print(df.head())
print("\n\n\nFind  All Columns Which All Entries Present :")

print(df.loc[:, df.notnull().all()])
print("\n\n\nRows & Columns Has A NaN:")

print(df.loc[:,df.isnull().any()])
print("\n\n\nDrop Rows With Any NaNs:")

print(df.dropna(how='any'))
```
