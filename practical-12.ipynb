## Practical No.12: Write a Python program using Scikit-learn to print the keys, number of rows-columns, feature names and the description of the given data.

```python
# Name : Kishan B. Lashkari
# Enrollmenr : 216270307002

from sklearn import datasets

iris = datasets.load_wine()
#df = datasets.read_csv('/kaggle/input/nifty50-stock-market-data/RELIANCE.csv')

print("Keys :")
print(iris.keys())

n_rows, n_cols = iris.data.shape
print(f"\n\nNumber of Rows :{n_rows}, Number of columns: {n_cols}")

print("\n\nFeature Names :")
print(iris.feature_names)

print("\n\nDescription :")
print(iris.DESCR)
```
